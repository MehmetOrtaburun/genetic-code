```python
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors
import numpy as np
import random

b_yesil=plt.plot([3], [8], 'seagreen',marker='o', markersize=70)
b_mavi=plt.plot([7], [8], 'darkturquoise',marker='o', markersize=70)
k_kırmızı=plt.plot([2], [4], 'r',marker='o', markersize=30)
k_sarı=plt.plot([3], [2], 'yellow',marker='o', markersize=30)
k_mavi=plt.plot([5], [1], 'royalblue',marker='o', markersize=30)
k_bordo=plt.plot([7], [2], 'darkred',marker='o', markersize=30)
k_turuncu=plt.plot([8], [4], 'orange',marker='o', markersize=30)
plt.axis((0,10,0,10))
plt.show()

goz_yesil=[3,8,7]
goz_darkturquoise=[7,8,7]
kirmizi_top=[2,4,3]
sari_top=[3,2,3]
mavi_top=[5,1,3]
bordo_top=[7,2,3]
turuncu_top=[8,4,3]

yuz=np.array(goz_yesil+goz_darkturquoise+kirmizi_top+sari_top+mavi_top+bordo_top+turuncu_top)
```

```python
def yuz_olustur(yuz):
    plt.plot(yuz[0], yuz[1], 'seagreen',marker='o', markersize=yuz[2]*10)
    plt.plot(yuz[3], yuz[4], 'darkturquoise',marker='o', markersize=yuz[5]*10)
    plt.plot(yuz[6], yuz[7], 'r',marker='o', markersize=yuz[8]*10)
    plt.plot(yuz[9], yuz[10], 'yellow',marker='o', markersize=yuz[11]*10)
    plt.plot(yuz[12], yuz[13], 'royalblue',marker='o', markersize=yuz[14]*10)
    plt.plot(yuz[15], yuz[16], 'darkred',marker='o', markersize=yuz[17]*10)
    plt.plot(yuz[18], yuz[19], 'orange',marker='o', markersize=yuz[20]*10)
    plt.axis((0,10,0,10))
    plt.show()
    
```

```python
iterasyon=600
crossover_rate=0.50
pop_size=10
gen_size=21

def random_gen():
    gen=random.randint(1,9)
    return gen

def create_chromosome():
    return [random_gen() for x in range(0,gen_size)]

def create_initial_population():
    return [create_chromosome() for x in range(0,pop_size)]

def fitness(birey):
    return 1/ (1 + sum((yuz - birey)**2)) 

def probability(fitness_values):
    P=[]
    total=sum(fitness_values)
    for f in fitness_values:
        P.append(f/total)
    return P

def crossover(p1,p2):
    o1=[]
    o2=[]
    c=random.randint(1,gen_size-1)
    print("Cut point:",c)
    o1[:c]=p2[:c]
    o1[c:]=p1[c:]
    
    o2[:c]=p1[:c]
    o2[c:]=p2[c:]
    print(o1)
    print(o2)
    return o1,o2

def mutasyon(mut):
    temp=[]
    temp=mut[:]
    gen=random_gen()
    index=random.randint(0,gen_size-1)
    temp[index]=gen
    return temp

```

```python
# Populasyon oluştur, fitness değerlerini hesapla listele
population=create_initial_population()
fitness_values=[]
for i in population:
    print(i," ",fitness(i))
    fitness_values.append(fitness(i))
epok=0
while(epok<iterasyon):
    P=probability(fitness_values) #Tüm bireylerin olasılıklarını hesapla 
    print("***************")
    j=0
    for i in population:
        print(i," ",fitness(i)," ",P[j])
        j+=1
    C=np.cumsum(P)# kümülatif toplamını bul
    print(C)

    rulet_parents=[]
    for i in range(0,len(C)):
        r=random.random()
        for j in range(0,len(C)):
            if C[j]>r:
                rulet_parents.append(j)# crossover için seçilecek bireylerin indisleri
                break

    print(rulet_parents)


    crossover_parents=[]
    k=0
    while(k<pop_size):
        r=random.random()
        if(r<crossover_rate):
            if(rulet_parents[k] not in crossover_parents):
                crossover_parents.append(rulet_parents[k])
        k=k+1
    print("Çaprazlanacak bireyler:",crossover_parents)

   

    if(len(crossover_parents)>1):
        for i in range(0,len(crossover_parents)):
            for j in range(i+1,len(crossover_parents)):
                
                # Çaprazlanacak bireyleri seç 
                o1,o2=crossover(population[crossover_parents[i]],population[crossover_parents[j]])
                print(crossover_parents[i]," ",crossover_parents[j])
                
                # Çaprazlanan bireyleri populasyona ekle
                population.append(o1)
                population.append(o2)
                
                # yeni gelen bireylerin uygunluklarını hesapla
                fitness_values.append(fitness(o1))
                fitness_values.append(fitness(o2))
    else:
        print("Caprazlamadan yeteri birey gelmedi!!")

    print(population)
    print(fitness_values)

    #  populasyondan seçilen bir bireyin bir genini değiştir.
    mut=mutasyon(population[random.randint(0,len(population)-1)]) 

    # mutasyondan gelen bireyi populasyona ekle ve uygunluğunu hesapla 
    population.append(mut)
    fitness_values.append(fitness(mut))

    #zip fonksiyonu ile her bireye ait fitnes değerini bireyi ile eşle listeye ekle 
    zip_list=zip(fitness_values,population)

    # listelenen bireyleri en iyiden kötüye doğru sırala 
    sort_list=sorted(zip_list,reverse=True)

    for j,i in list(sort_list):
        print(i," ",j)

    p=len(population)
    while(p>pop_size):
        sort_list.pop()
        p=p-1
    print("*****************")
    for j,i in list(sort_list):
        print(i," ",j)

    population=[]
    fitness_values=[]

    # sıralı listeden en ugun olanları yeni populasyon için listeye ekle 
    for j,i in list(sort_list):
        population.append(i)
        fitness_values.append(j)
    epok=epok+1

print("Son poplasyon:",population)
print("Uygunluk degerleri:",fitness_values)
```
